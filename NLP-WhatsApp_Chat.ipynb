```python
import pandas as pd
```

```python
df = pd.read_csv("mabel.txt",header=None,on_bad_lines='skip',encoding='utf8')
```

# Preprocessing

```python
df=df.drop(0)
df.columns=['Date','Chat']
Message=df["Chat"].str.split("-",n=1,expand=True)
df["Time"]=Message[0]
Message1=Message[1].str.split(":",n=1,expand=True)
df["Name"]=Message1[0]
df["Chat"]=Message1[1]
df=df[["Date","Time","Name","Chat"]]
df
```

# SENTIMENTAL ANALYSIS

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
dataset=df
dataset.dropna(inplace=True)
```

```python
senti=sid.polarity_scores(dataset['Chat'][4])
senti
```

```python
 def sentimentalAnalysis(data,columnname):
        
        #download vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

```python
posneg,data=sentimentalAnalysis(dataset,columnname='Chat')
```

# Topic Modeling

# Preprocessing (text to matrix(vector))

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')  # to ignore common words like(ex:"is","was","the","an" etc...)
dtm=tfidf.fit_transform(df["Chat"])
```

# Model_Creation

```python
from sklearn.decomposition import NMF
nmf_model=NMF(n_components=5,random_state=42) # "n_components" means ,take 5 data from overall data and create model using 5 datas...
nmf_model.fit(dtm)
```

```python
#### it is just for viewing "5 data" from our created model by "revoking" a preprocessing technique...####  

for index,topic in enumerate(nmf_model.components_):
    results=([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]]) # revoking_preprocessing("matrix(vector)" to "text")....
    print(results)
```

```python
topic_results=nmf_model.transform(dtm) # in model creation above we just "fit" dtm in a model,so here we are transforming...
df["Topic"]=topic_results.argmax(axis=1) # "argmax()" gives "index" of "max values"
df
```
